### 01 - 02 - Testar os três browsers na mesma execução

```python
# Testar os três browsers na mesma execução
# Gera três screenshots

# Importa a API síncrona do Playwright
from playwright.sync_api import sync_playwright

# Inicia o Playwright
with sync_playwright() as p:

    # Cria uma lista com os três browsers disponíveis
    # Cada item é um objeto de tipo BrowserType
    browsers_disponiveis = [
        p.chromium,   # Google Chrome / Edge
        p.firefox,    # Mozilla Firefox
        p.webkit,     # Webkit
    ]

    # Percorre cada browser da lista
    for browser_type in browsers_disponiveis:

        # Imprime qual browser está sendo testado
        # browser_type.name retorna 'chromium', 'firefox' ou 'webkit'
        print(f'\nTestando: {browser_type.name}')

        # Abre o browser atual sem interface gráfica
        browser = browser_type.launch(headless=True)

        # Abre uma nova página
        page = browser.new_page()

        # Navega para o site de testes
        page.goto('https://www.saucedemo.com', wait_until='networkidle')

        # Captura o título da página
        titulo = page.title()

        # Imprime o resultado
        print(f'  Título: {titulo}')
        print(f'  URL: {page.url}')

        # Define o nome do arquivo usando o nome do browser
        nome_arquivo = f'screenshot_{browser_type.name}.png'

        # Tira screenshot com nome específico de cada browser
        page.screenshot(path=nome_arquivo)
        print(f'  Screenshot salvo: {nome_arquivo}')

        # Fecha o browser antes de abrir o próximo
        browser.close()

    print('\nTodos os browsers testados com sucesso!')
```

### 02 - 01 - Comparar headless vs headed

```python
# Comparar headless vs headed

# Importa a API síncrona do Playwright
from playwright.sync_api import sync_playwright

# Importa datetime para medir o tempo de cada modo
from datetime import datetime

with sync_playwright() as p:

    # Modo headless

    print('Iniciando modo HEADLESS (sem janela)...')

    # Registra o tempo de início do modo headless
    inicio_headless = datetime.now()

    # Abre o browser SEM interface gráfica
    # headless=True é o padrão — aqui deixamos explícito para clareza
    browser_headless = p.chromium.launch(headless=True)

    # Abre uma nova página
    page_headless = browser_headless.new_page()

    # Navega para o site
    page_headless.goto('https://www.saucedemo.com')

    # Captura o título
    titulo = page_headless.title()

    # Calcula o tempo gasto
    tempo_headless = (datetime.now() - inicio_headless).total_seconds()

    print(f'  Título: {titulo}')
    print(f'  Tempo: {tempo_headless:.2f} segundos')

    # Fecha o browser headless
    browser_headless.close()

    # Modo headed

    print('\nIniciando modo HEADED (com janela)...')

    # Registra o tempo de início do modo headed
    inicio_headed = datetime.now()

    # Abre o browser COM interface gráfica
    # headless=False abre a janela do browser na tela
    browser_headed = p.chromium.launch(headless=False)

    # Abre uma nova página
    page_headed = browser_headed.new_page()

    # Navega para o site — você verá a janela abrir
    page_headed.goto('https://www.saucedemo.com')

    # Captura o título
    titulo = page_headed.title()

    # Calcula o tempo gasto
    tempo_headed = (datetime.now() - inicio_headed).total_seconds()

    print(f'  Título: {titulo}')
    print(f'  Tempo: {tempo_headed:.2f} segundos')

    # Fecha o browser headed
    browser_headed.close()

    # Comparação

    print(f'\nHeadless: {tempo_headless:.2f}s')
    print(f'Headed:   {tempo_headed:.2f}s')
    print(f'Diferença: {abs(tempo_headed - tempo_headless):.2f}s')
```

### 03 - 01 - Sync: acessar três páginas em sequência

```python
# Sync: acessar três páginas em sequência

# Importa a API síncrona do Playwright
from playwright.sync_api import sync_playwright

# Importa datetime para medir o tempo total
from datetime import datetime

# Registra o tempo de início
inicio = datetime.now()

# Lista de URLs para acessar em sequência
urls = [
    'https://www.saucedemo.com',
    'https://www.selenium.dev/selenium/web/web-form.html',
    'https://the-internet.herokuapp.com',
]

with sync_playwright() as p:

    # Abre o browser
    browser = p.chromium.launch(headless=True)

    # Abre uma página — será reutilizada para cada URL
    page = browser.new_page()

    # Percorre cada URL da lista
    for url in urls:

        # Navega para a URL atual
        # No modo sync, o script PARA aqui até a página carregar
        page.goto(url)

        # Aguarda a rede ficar quieta
        page.wait_for_load_state('networkidle')

        # Captura o título da página
        titulo = page.title()

        # Imprime o resultado
        print(f'URL: {url}')
        print(f'Título: {titulo}')
        print()

    # Fecha o browser
    browser.close()

# Calcula o tempo total
total = (datetime.now() - inicio).total_seconds()
print(f'Tempo total (sequencial): {total:.2f} segundos')
```

### 03 - 02 - Async: acessar três páginas em paralelo

```python
# Async: acessar três páginas em paralelo
# Importa o módulo asyncio para gerenciar a execução assíncrona
import asyncio

# Importa a API ASSÍNCRONA do Playwright
# Note: async_playwright em vez de sync_playwright
from playwright.async_api import async_playwright

# Importa datetime para medir o tempo total
from datetime import datetime

# Define a função que acessa UMA página — note o async def
async def acessar_pagina(browser, url):

    # Abre uma nova página dentro do browser
    # O await libera o controle enquanto espera a página abrir
    page = await browser.new_page()

    # Navega para a URL
    # O await libera o controle enquanto a página carrega
    # Outras páginas podem carregar ao mesmo tempo
    await page.goto(url)

    # Aguarda a rede ficar quieta
    await page.wait_for_load_state('networkidle')

    # Captura o título
    titulo = await page.title()

    # Fecha a página após capturar o título
    await page.close()

    # Retorna o resultado como tupla
    return url, titulo

# Define a função principal — também precisa ser async
async def acessar_paralelo():

    # Registra o tempo de início
    inicio = datetime.now()

    # Lista de URLs para acessar em paralelo
    urls = [
        'https://www.saucedemo.com',
        'https://www.selenium.dev/selenium/web/web-form.html',
        'https://the-internet.herokuapp.com',
    ]

    # Inicia o Playwright assíncrono
    async with async_playwright() as p:

        # Abre o browser
        browser = await p.chromium.launch(headless=True)

        # Cria uma lista de tarefas — uma para cada URL
        # Cada tarefa é uma chamada a acessar_pagina()
        # mas NENHUMA executa ainda — são apenas agendadas
        tarefas = [
            acessar_pagina(browser, url)
            for url in urls
        ]

        # asyncio.gather() dispara TODAS as tarefas ao mesmo tempo
        # e aguarda todas terminarem
        # É aqui que o paralelismo acontece de fato
        resultados = await asyncio.gather(*tarefas)

        # Percorre e imprime os resultados
        for url, titulo in resultados:
            print(f'URL: {url}')
            print(f'Título: {titulo}')
            print()

        # Fecha o browser
        await browser.close()

    # Calcula o tempo total
    total = (datetime.now() - inicio).total_seconds()
    print(f'Tempo total (paralelo): {total:.2f} segundos')

# asyncio.run() inicia o loop assíncrono e executa a função principal
asyncio.run(acessar_paralelo())
```

### 04 - 01 - Comparar todas as formas de localizar o mesmo elemento

```python
# Comparar todas as formas de localizar o mesmo elemento
# Importa a API síncrona do Playwright
from playwright.sync_api import sync_playwright

with sync_playwright() as p:

    browser = p.chromium.launch(headless=False, slow_mo=600)
    page = browser.new_page()
    page.goto('https://www.saucedemo.com')
    page.wait_for_load_state('networkidle')

    print('Testando diferentes formas de localizar o campo Username:\n')

    # ── Forma 1: get_by_placeholder ────────────────────────────────────
    # Localiza pelo texto do placeholder do input
    locator1 = page.get_by_placeholder('Username')
    locator1.fill('standard_user')
    print('1. get_by_placeholder: OK')
    locator1.clear()

    # ── Forma 2: get_by_test_id (requer configuração) ──────────────────
    # Configura o Playwright para usar data-test como atributo de teste
    p.selectors.set_test_id_attribute('data-test')
    locator2 = page.get_by_test_id('username')
    locator2.fill('standard_user')
    print('2. get_by_test_id: OK')
    locator2.clear()

    # ── Forma 3: CSS por id ────────────────────────────────────────────
    # O # indica seleção por id no CSS
    locator3 = page.locator('#user-name')
    locator3.fill('standard_user')
    print('3. CSS por id: OK')
    locator3.clear()

    # ── Forma 4: CSS por atributo data-test ────────────────────────────
    # Seletor CSS que busca pelo atributo data-test
    locator4 = page.locator('[data-test="username"]')
    locator4.fill('standard_user')
    print('4. CSS por atributo data-test: OK')
    locator4.clear()

    # ── Forma 5: CSS por name ──────────────────────────────────────────
    # Seletor CSS que busca pelo atributo name
    locator5 = page.locator('[name="user-name"]')
    locator5.fill('standard_user')
    print('5. CSS por name: OK')
    locator5.clear()

    # ── Forma 6: XPath ─────────────────────────────────────────────────
    # XPath que busca input com id igual a user-name
    locator6 = page.locator("//input[@id='user-name']")
    locator6.fill('standard_user')
    print('6. XPath: OK')
    locator6.clear()

    print('\nTodas as formas funcionaram!')
    browser.close()
```

### 04 - 02 - Trabalhar com listas de elementos

```python
# Trabalhar com listas de elementos
# Importa a API síncrona do Playwright
from playwright.sync_api import sync_playwright

with sync_playwright() as p:

    browser = p.chromium.launch(headless=False)
    page = browser.new_page()

    # Faz login no saucedemo
    page.goto('https://www.saucedemo.com')
    page.get_by_placeholder('Username').fill('standard_user')
    page.get_by_placeholder('Password').fill('secret_sauce')
    page.get_by_role('button', name='Login').click()
    page.wait_for_load_state('networkidle')

    # Contar elementos

    # Localiza todos os nomes de produtos
    locator_nomes = page.locator('[data-test="inventory-item-name"]')

    # count() retorna o número total de elementos encontrados
    total = locator_nomes.count()
    print(f'Total de produtos: {total}')

    # Pegar o primeiro e o último

    # first acessa o primeiro elemento da lista
    primeiro = locator_nomes.first.text_content()
    print(f'Primeiro produto: {primeiro}')

    # last acessa o último elemento da lista
    ultimo = locator_nomes.last.text_content()
    print(f'Último produto: {ultimo}')

    # Pegar por índice

    # nth(índice) acessa o elemento na posição indicada
    # índices começam em 0
    terceiro = locator_nomes.nth(2).text_content()
    print(f'Terceiro produto: {terceiro}')

    # Pegar todos de uma vez

    # all_text_contents() retorna uma lista Python com o texto de cada elemento
    todos = locator_nomes.all_text_contents()
    print('\nTodos os produtos:')
    for i, nome in enumerate(todos, start=1):
        print(f'  {i}. {nome}')

    # Filtrar por texto

    # filter(has_text=) retorna apenas os elementos que contêm o texto
    locator_filtrado = page.locator('[data-test="inventory-item-name"]')
    locator_filtrado = locator_filtrado.filter(has_text='Sauce Labs')
    total_sauce = locator_filtrado.count()
    print(f'\nProdutos com "Sauce Labs" no nome: {total_sauce}')

    browser.close()
```

### 05 -01 - Seletores CSS variados

```python
# Seletores CSS variados
# Importa a API síncrona do Playwright
from playwright.sync_api import sync_playwright

with sync_playwright() as p:

    browser = p.chromium.launch(headless=False, slow_mo=500)
    page = browser.new_page()
    page.goto('https://www.saucedemo.com')
    page.wait_for_load_state('networkidle')

    # Seletor por tag
    # Seleciona TODOS os inputs da página
    # Retorna uma coleção — usamos .count() para ver quantos são
    total_inputs = page.locator('input').count()
    print(f'Total de inputs na página: {total_inputs}')

    # Seletor por classe
    # O ponto (.) indica seleção por classe CSS
    # form_input é a classe dos campos de login
    total_form_inputs = page.locator('.form_input').count()
    print(f'Inputs com classe form_input: {total_form_inputs}')

    # Seletor por id
    # O # indica seleção por id
    # Seleciona especificamente o campo user-name
    campo_usuario = page.locator('#user-name')
    campo_usuario.fill('standard_user')
    print('Campo usuário preenchido via id: OK')

    # Seletor por atributo
    # Colchetes [] indicam seleção por atributo
    # Seleciona input cujo type é password
    campo_senha = page.locator('input[type="password"]')
    campo_senha.fill('secret_sauce')
    print('Campo senha preenchido via atributo type: OK')

    # Seletor combinado: tag + atributo
    # Seleciona input que tem type submit E name login-button
    botao = page.locator('input[type="submit"][name="login-button"]')
    botao.click()
    page.wait_for_load_state('networkidle')
    print('Login realizado via seletor combinado: OK')

    # Seletor por classe após login
    # Conta quantos itens de inventário existem na página de produtos
    total_itens = page.locator('.inventory_item').count()
    print(f'Total de itens no inventário: {total_itens}')

    browser.close()
```

### 05 -02 - XPath com diferentes estratégias

```python
# XPath com diferentes estratégias
# Importa a API síncrona do Playwright
from playwright.sync_api import sync_playwright

with sync_playwright() as p:

    browser = p.chromium.launch(headless=False, slow_mo=500)
    page = browser.new_page()
    page.goto('https://www.saucedemo.com')
    page.wait_for_load_state('networkidle')

    # XPath por atributo id
    # // significa "em qualquer lugar da página"
    # input[@id='user-name'] significa "input com id igual a user-name"
    page.locator("//input[@id='user-name']").fill('standard_user')
    print('1. XPath por id: OK')

    # XPath por atributo type
    # Seleciona o input cujo type é password
    page.locator("//input[@type='password']").fill('secret_sauce')
    print('2. XPath por type: OK')

    # XPath por atributo name
    # Seleciona o input cujo name é login-button
    page.locator("//input[@name='login-button']").click()
    page.wait_for_load_state('networkidle')
    print('3. XPath por name: OK')

    # XPath por texto
    # text() captura o texto interno do elemento
    # Seleciona o elemento div que contém exatamente o texto indicado
    produto = page.locator(
        "//div[text()='Sauce Labs Backpack']"
    ).text_content()
    print(f'4. XPath por texto: {produto}')

    # XPath por texto parcial
    # contains() verifica se o texto CONTÉM o valor indicado
    # útil quando o texto completo é longo ou variável
    preco = page.locator(
        "//div[contains(@class,'inventory_item_price')]"
    ).first.text_content()
    print(f'5. XPath por classe parcial: {preco}')

    browser.close()
```

### 05 - 03 - Localizadores Semânticos em Formulário

```python
# Localizadores Semânticos em Formulário
# Importa a API síncrona do Playwright
from playwright.sync_api import sync_playwright

with sync_playwright() as p:

    # Abre o Chromium com interface gráfica
    # slow_mo=700 para acompanhar cada passo com clareza em aula
    browser = p.chromium.launch(headless=False, slow_mo=700)

    # Abre uma nova página
    page = browser.new_page()

    # Navega para o formulário de testes
    page.goto('https://www.selenium.dev/selenium/web/web-form.html')

    # Aguarda o formulário carregar
    page.wait_for_load_state('domcontentloaded')

    # get_by_label
    # Localiza o campo pela sua label — o texto descritivo ao lado
    # ou acima do campo no formulário
    # No HTML:
    #   <label for="my-text-id">Text input</label>
    #   <input id="my-text-id" type="text" ...>
    # O Playwright conecta automaticamente a label ao input pelo atributo for/id
    page.get_by_label('Text input').fill('Automacao com Playwright')
    print('get_by_label("Text input"): preenchido')

    # Localiza o campo de senha pela sua label
    # No HTML:
    #   <label for="my-password-id">Password</label>
    #   <input id="my-password-id" type="password" ...>
    page.get_by_label('Password').fill('senha_segura_123')
    print('get_by_label("Password"): preenchido')

    # Localiza o textarea pela sua label
    # No HTML:
    #   <label for="my-textarea-id">Textarea</label>
    #   <textarea id="my-textarea-id" ...>
    page.get_by_label('Textarea').fill('Texto preenchido automaticamente.')
    print('get_by_label("Textarea"): preenchido')

    # get_by_label — select
    # get_by_label também funciona com campos <select>
    # No HTML:
    #   <label for="my-select-id">Dropdown (select)</label>
    #   <select id="my-select-id">
    #     <option value="1">One</option>
    #     <option value="2">Two</option>
    #     <option value="3">Three</option>
    #   </select>
    page.get_by_label('Dropdown (select)').select_option('Two')
    print('get_by_label("Dropdown (select)"): opção Two selecionada')

    # get_by_label — checkbox
    # get_by_label localiza checkboxes pelo texto da label associada
    # No HTML:
    #   <input type="checkbox" id="my-check-id">
    #   <label for="my-check-id">Default checkbox</label>
    page.get_by_label('Default checkbox').check()

    # is_checked() retorna True se o checkbox está marcado
    marcado = page.get_by_label('Default checkbox').is_checked()
    print(f'get_by_label("Default checkbox"): marcado = {marcado}')

    # get_by_label — radio
    # O mesmo princípio funciona para radio buttons
    # No HTML:
    #   <input type="radio" id="my-radio-id">
    #   <label for="my-radio-id">Default radio</label>
    page.get_by_label('Default radio').check()
    print('get_by_label("Default radio"): selecionado')

    # get_by_label — date picker
    # Localiza o campo de data pela sua label
    # O formato aceito é YYYY-MM-DD
    page.get_by_label('Date picker').fill('2025-06-15')
    print('get_by_label("Date picker"): data preenchida')

    # get_by_role — button
    # Localiza o botão de envio pelo papel semântico e texto visível
    # role='button' — papel semântico do elemento
    # name='Submit' — texto visível do botão
    page.get_by_role('button', name='Submit').click()
    print('get_by_role("button", name="Submit"): clicado')

    # Aguarda a página de confirmação carregar
    page.wait_for_load_state('domcontentloaded')

    # get_by_text — mensagem de confirmação
    # Após o envio, o site exibe uma mensagem de confirmação
    # get_by_text localiza qualquer elemento pelo seu texto visível
    # exact=True (padrão) exige correspondência exata do texto
    mensagem = page.get_by_text('Received!')

    # Verifica se a mensagem está visível na página
    visivel = mensagem.is_visible()
    print(f'get_by_text("Received!"): visível = {visivel}')

    # Captura o texto completo do elemento
    texto = mensagem.text_content()
    print(f'Mensagem capturada: {texto}')

    # Tira screenshot da confirmação como evidência
    page.screenshot(path='semanticos_formulario.png')
    print('\nScreenshot salvo: semanticos_formulario.png')

    # Fecha o browser
    browser.close()
```

### 06 -02 - Atalhos de teclado e teclas especiais

```python
# Atalhos de teclado e teclas especiais
# Importa a API síncrona do Playwright
from playwright.sync_api import sync_playwright
import time

with sync_playwright() as p:

    browser = p.chromium.launch(headless=False, slow_mo=500)
    page = browser.new_page()
    page.goto('https://www.selenium.dev/selenium/web/web-form.html')
    page.wait_for_load_state('domcontentloaded')

    # Preenche o campo de texto
    campo = page.get_by_label('Text input')
    campo.fill('Texto de exemplo para demonstração')

    # End move o cursor para o final da linha
    campo.press('End')
    print('End: cursor foi para o final')
    time.sleep(2)

    # Home move o cursor para o início da linha
    campo.press('Home')
    print('Home: cursor voltou para o início')
    time.sleep(2)

    # Shift + ArrowRight seleciona um caractere à direita
    # Pressiona Shift e mantém enquanto pressiona ArrowRight 5 vezes
    page.keyboard.down('Shift')
    for _ in range(5):
        # Seleciona um caractere a cada pressão
        page.keyboard.press('ArrowRight')
    page.keyboard.up('Shift')
    print('Shift + ArrowRight x5: 5 caracteres selecionados')
    time.sleep(2)

    # Digita para substituir o texto selecionado
    page.keyboard.type('NOVO ')
    print('Texto selecionado substituído por "NOVO "')
    time.sleep(2)

    # Backspace apaga o caractere ANTES do cursor
    campo.press('Backspace')
    print('Backspace: caractere antes do cursor apagado')
    time.sleep(2)

    # Shift+Home seleciona tudo desde o cursor até o início da linha
    campo.press('Shift+Home')
    print('Shift+Home: selecionou do cursor até o início')
    time.sleep(2)

    # Pressiona Delete para apagar a seleção
    campo.press('Delete')
    print('Delete: seleção apagada')
    time.sleep(2)

    # Preenche o campo com o conteúdo final usando fill()
    campo.fill('Conteúdo final após manipulações de teclado')
    time.sleep(2)

    # Tira screenshot do resultado
    page.screenshot(path='teclas_especiais.png')
    print('\nScreenshot salvo: teclas_especiais.png')

    browser.close()
```

```python
# Diferentes tipos de clique
# Importa a API síncrona do Playwright
from playwright.sync_api import sync_playwright
import time

with sync_playwright() as p:

    browser = p.chromium.launch(headless=False, slow_mo=600)
    page = browser.new_page()

    # Faz login no saucedemo
    page.goto('https://www.saucedemo.com')
    page.get_by_placeholder('Username').fill('standard_user')
    page.get_by_placeholder('Password').fill('secret_sauce')
    page.get_by_role('button', name='Login').click()
    page.wait_for_load_state('networkidle')

    # Localiza o primeiro botão "Add to cart"
    botao_add = page.locator('[data-test="add-to-cart-sauce-labs-backpack"]')

    # click() executa um clique simples com o botão esquerdo
    botao_add.click()
    print('Clique simples: produto adicionado ao carrinho')
    time.sleep(2)

    # Após o clique o botão muda de texto para "Remove"
    # text_content() captura o texto atual do botão
    botao_remove = page.locator('[data-test="remove-sauce-labs-backpack"]')
    texto = botao_remove.text_content()
    print(f'Texto do botão após clique: {texto}')
    time.sleep(2)

    # Localiza o ícone do carrinho no topo da página
    carrinho = page.locator('.shopping_cart_link')

    # hover() move o mouse até o elemento sem clicar
    # Útil para revelar tooltips ou menus que aparecem ao passar o mouse
    carrinho.hover()
    print('Hover no carrinho: mouse posicionado')
    time.sleep(2)

    # Agora clica no carrinho após o hover
    carrinho.click()
    page.wait_for_load_state('networkidle')
    print(f'Clique no carrinho: URL atual = {page.url}')
    time.sleep(2)

    # Volta para a lista de produtos
    page.go_back()
    page.wait_for_load_state('networkidle')

    # Localiza o container do segundo produto
    segundo_produto = page.locator('.inventory_item').nth(1)

    # click() com position clica em coordenadas relativas ao elemento
    # x=10, y=10 clica no canto superior esquerdo do elemento
    segundo_produto.click(position={'x': 10, 'y': 10})
    print('Clique com coordenada relativa: OK')
    time.sleep(2)

    # Tira screenshot
    page.screenshot(path='cliques.png')
    print('Screenshot salvo: cliques.png')
    time.sleep(2)

    browser.close()
```

### 07 - 02 - Scroll e drag and drop

```python
# Scroll e drag and drop
# Importa a API síncrona do Playwright
from playwright.sync_api import sync_playwright
import time

with sync_playwright() as p:

    browser = p.chromium.launch(headless=False, slow_mo=500)
    page = browser.new_page()

    # Navega para o site the-internet que tem exemplos de drag and drop
    page.goto('https://the-internet.herokuapp.com/drag_and_drop')
    page.wait_for_load_state('domcontentloaded')

    # Captura o texto do elemento A antes do drag
    # O elemento A tem id="column-a"
    texto_a_antes = page.locator('#column-a header').text_content()
    texto_b_antes = page.locator('#column-b header').text_content()
    print(f'Antes — A: {texto_a_antes} | B: {texto_b_antes}')
    time.sleep(2)

    # drag_to() executa o arrastar e soltar automaticamente
    # Arrasta o elemento A para a posição do elemento B
    page.locator('#column-a').drag_to(page.locator('#column-b'))
    print('Drag and drop executado')
    time.sleep(2)

    # Verifica se as posições foram trocadas
    texto_a_depois = page.locator('#column-a header').text_content()
    texto_b_depois = page.locator('#column-b header').text_content()
    print(f'Depois — A: {texto_a_depois} | B: {texto_b_depois}')
    time.sleep(2)

    # Tira screenshot do resultado
    page.screenshot(path='drag_drop.png')
    print('Screenshot salvo: drag_drop.png')
    time.sleep(2)

    # Navega para uma página com mais conteúdo para demonstrar scroll
    page.goto('https://the-internet.herokuapp.com')
    page.wait_for_load_state('domcontentloaded')

    # Captura a posição atual do scroll
    # window.scrollY retorna quantos pixels foram rolados verticalmente
    pos_inicial = page.evaluate('window.scrollY')
    print(f'\nPosição inicial do scroll: {pos_inicial}px')
    time.sleep(2)

    # mouse.wheel() rola a página
    # Primeiro parâmetro: delta horizontal (0 = sem rolagem horizontal)
    # Segundo parâmetro: delta vertical (500 = rola 500px para baixo)
    page.mouse.wheel(0, 500)
    print('Rolou 500px para baixo')
    time.sleep(2)

    # Captura a nova posição do scroll
    pos_depois = page.evaluate('window.scrollY')
    print(f'Posição após scroll: {pos_depois}px')
    time.sleep(2)

    # Rola até o final da página usando JavaScript
    page.evaluate('window.scrollTo(0, document.body.scrollHeight)')
    print('Rolou até o final da página')
    time.sleep(2)

    # Tira screenshot do final da página
    page.screenshot(path='scroll_final.png')
    print('Screenshot salvo: scroll_final.png')
    time.sleep(2)

    browser.close()
```
